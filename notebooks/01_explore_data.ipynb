```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV you created
df = pd.read_csv('../data/raw/all_articles.csv')

print(f"Total articles: {len(df)}")
print(f"\nFirst few rows:")
df.head()
```

```python
print("Missing values:")
print(df.isnull().sum())

print("\nData types:")
print(df.dtypes)
```

```python
plt.figure(figsize=(10, 6))
df['source'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 News Sources')
plt.xlabel('Source')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['text_length'] = df['text'].str.len()

plt.figure(figsize=(10, 6))
plt.hist(df['text_length'], bins=50)
plt.title('Distribution of Article Lengths')
plt.xlabel('Number of Characters')
plt.ylabel('Frequency')
plt.show()

print(f"Average article length: {df['text_length'].mean():.0f} characters")
```

```python
# Cell 5: Word cloud (most common words)
from wordcloud import WordCloud

# Combine all text
all_text = ' '.join(df['text'].dropna())

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Articles')
plt.show()
```
